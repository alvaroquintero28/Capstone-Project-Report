```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/Users/alvaroquintero/Project/Capstone-Project-Report/Cleaned Data/cleaned_stroke_prediction_dataset.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the data to understand its structure
print("First few rows of the dataset:")
print(data.head())

# Display basic information about the dataset
print("\nDataset Information:")
print(data.info())

# Summary statistics of the dataset
summary_statistics = data.describe(include='all')
print("\nSummary Statistics:")
print(summary_statistics)

# Check for missing values
missing_values = data.isnull().sum()
print("\nMissing Values in Each Column:")
print(missing_values[missing_values > 0])  # Only display columns with missing values

# Correlation analysis
correlation_matrix = data.corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)

# Set the style for the plots
sns.set(style="whitegrid")

# Create a figure for multiple plots
plt.figure(figsize=(15, 12))

# Plot distributions of key features
features_to_plot = ['age', 'hypertension', 'heart_disease', 'ever_married', 'bmi', 'smoking_status']

for i, feature in enumerate(features_to_plot, start=1):
    plt.subplot(3, 2, i)
    sns.histplot(data, x=feature, hue='stroke', multiple='stack', kde=True, bins=20)
    plt.title(f'Distribution of {feature} by Stroke')
    plt.xlabel(feature)
    plt.ylabel('Count')

plt.tight_layout()
plt.show()

# Visualizing the relationship between two key features (e.g., Age and Stroke)
plt.figure(figsize=(10, 6))
sns.boxplot(x='stroke', y='age', data=data)
plt.title('Boxplot of Age by Stroke Status')
plt.xlabel('Stroke')
plt.ylabel('Age')
plt.xticks(ticks=[0, 1], labels=['No Stroke', 'Stroke'])
plt.show()

# Visualizing categorical relationships with count plots
plt.figure(figsize=(14, 6))
sns.countplot(x='work_type', hue='stroke', data=data)
plt.title('Count of Stroke by Work Type')
plt.xlabel('Work Type')
plt.ylabel('Count')
plt.legend(title='Stroke Status', labels=['No Stroke', 'Stroke'])
plt.show()

# Bar plot of smoking status effect on stroke
plt.figure(figsize=(10, 6))
sns.countplot(x='smoking_status', hue='stroke', data=data)
plt.title('Count of Stroke by Smoking Status')
plt.xlabel('Smoking Status')
plt.ylabel('Count')
plt.legend(title='Stroke Status', labels=['No Stroke', 'Stroke'])
plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the uploaded file
file_path = '/Users/alvaroquintero/Project/Capstone-Project-Report/Cleaned Data/cleaned_work_life_balance.csv'
data = pd.read_csv(file_path)

# Display basic information about the dataset
data_info = data.info()
data_head = data.head()

# Show dataset info and first few rows
print("Dataset Information:")
print(data_info)
print("\nFirst Few Rows of the Dataset:")
print(data_head)

# Summary statistics
summary_stats = data.describe(include='all')
print("\nSummary Statistics:")
print(summary_stats)

# Check for missing values
missing_values = data.isnull().sum()
print("\nMissing Values in Each Column:")
print(missing_values[missing_values > 0])  # Show only columns with missing values

# Visualizations for daily stress and sleep hours
plt.figure(figsize=(12, 6))

# Distribution of DAILY_STRESS
plt.subplot(1, 2, 1)
sns.histplot(data['DAILY_STRESS'], bins=10, kde=True)
plt.title('Distribution of Daily Stress')
plt.xlabel('Daily Stress Level')
plt.ylabel('Frequency')

# Boxplot of SLEEP_HOURS by GENDER
plt.subplot(1, 2, 2)
sns.boxplot(x='GENDER', y='SLEEP_HOURS', data=data)
plt.title('Sleep Hours by Gender')
plt.xlabel('Gender')
plt.ylabel('Sleep Hours')

plt.tight_layout()
plt.show()

# Examine data types
print("\nData Types in the Dataset:")
print(data.dtypes)

# Convert categorical columns to numeric as necessary
# Example: If the AGE column is categorical with ranges, create a mapping from category to numeric values
age_mapping = {
    '18 to 25': 21,
    '26 to 35': 30.5,
    '36 to 50': 43,
    '51 and above': 55
}

if 'AGE' in data.columns:
    data['AGE'] = data['AGE'].replace(age_mapping)

# Check if column 'SLEEP_HOURS' needs conversion (if it's categorical)
if data['SLEEP_HOURS'].dtype == 'object':
    print("\nConverting SLEEP_HOURS to numeric if necessary...")
    data['SLEEP_HOURS'] = pd.to_numeric(data['SLEEP_HOURS'], errors='coerce')

# Recalculate correlations after converting
correlation = data.select_dtypes(include=['float64', 'int64']).corr()

# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

# Additional visualizations
# Count plot for AGE and GENDER
plt.figure(figsize=(12, 6))
sns.countplot(x='AGE', hue='GENDER', data=data)
plt.title('Count of Gender by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.show()

# Violin plot for SLEEP_HOURS by GENDER
plt.figure(figsize=(12, 6))
sns.violinplot(x='GENDER', y='SLEEP_HOURS', data=data)
plt.title('Violin Plot of Sleep Hours by Gender')
plt.xlabel('Gender')
plt.ylabel('Sleep Hours')
plt.show()

# Pairplot for visualizing relationships among variables
sns.pairplot(data, hue='GENDER', diag_kind='kde')
plt.title('Pairplot of Key Features Colored by Gender')
plt.show()

# Time series analysis if the dataset contains a timestamp
if 'DATE' in data.columns:
    data['DATE'] = pd.to_datetime(data['DATE'])  # Convert to datetime
    plt.figure(figsize=(14, 7))
    sns.lineplot(x='DATE', y='DAILY_STRESS', hue='GENDER', data=data)
    plt.title('Daily Stress Over Time by Gender')
    plt.xlabel('Date')
    plt.ylabel('Daily Stress Level')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
